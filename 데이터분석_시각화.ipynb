# 막대 그래포

```
df.plot(kind='bar')
```
기본적으로 세로 형태이지만 가로로 하기 위해선

```
df.plot(kind='barh')
```

두가지이상을 스택으로 쌓기위해선

```
df.plot(kind='bar', stacked=True)
```
특정 값을 추출해서 하기위해선
```
df['추출컬럼'].plot(kind='bar')
```
